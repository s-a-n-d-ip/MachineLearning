```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import cross_val_score,KFold
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=1)
scores=cross_val_score(knn,X,y,cv=kfold,scoring='r2')
```

```python
scores.mean()
```

### Apply GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
knn=KNeighborsRegressor()
```

```python
grid={
    'n_neighbors':[1,3,5,7,9,11,12,15,20,31],
    'algorithm':['auto','ball_tree','kd_tree','brute'],
    'weights':['uniform','distance'],
    'p':[1,2,3,4]
}
```

```python
gsv=GridSearchCV(knn,grid,cv=kfold,scoring='r2',verbose=2,n_jobs=-1)
```

```python
gsv.fit(X,y)
```

```python
gsv.best_params_
```

```python
gsv.best_score_
```

```python
pd.DataFrame(gsv.cv_results_)
```

### RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
rsv=RandomizedSearchCV(knn,grid,cv=kfold,scoring='r2',verbose=2,n_jobs=-1)
```

```python
rsv.fit(X,y)
```

```python
rsv.best_params_
```

```python
rsv.best_score_
```
