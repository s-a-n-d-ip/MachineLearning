```python
from sklearn.datasets import load_diabetes
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y=load_diabetes(return_X_y=True)    
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.metrics import r2_score
y_pred=lr.predict(X_test)
r2_score(y_test,y_pred)
```

### Multiple Linear Regression using Gradient Descent from scratch (Batch Gradient Descent)

```python
class MLRGradientDescent:
    def __init__(self,learning_rate,epoch):
        self._coef=None
        self._intercept=None
        self.learning_rate=learning_rate
        self.epoch=epoch
        
    def fit(self,X_train,y_train):
        self._intercept=0
        self._coef=np.ones(X_train.shape[1]) # we initilize an array of ones with shape of X_train as there will be that many betas as columns in X_train 
        for i in range(self.epoch):
            y_pred=np.dot(X_train,self._coef)+self._intercept # X.Beta+intercept 
            slope_intercept=-2 * np.mean(y_train-y_pred)
            self._intercept=self._intercept-(self.learning_rate*slope_intercept)

            slope_coefs=-2*np.dot((y_train-y_pred),X_train)/X_train.shape[0]
            self._coef=self._coef-(self.learning_rate*slope_coefs)
        print(self._coef,self._intercept)
    def predict(self,X):
        return np.dot(X,self._coef)+self._intercept
```

```python
mlr=MLRGradientDescent(learning_rate=0.01,epoch=17000)
mlr.fit(X_train,y_train)
```

```python
r2_score(y_test,mlr.predict(X_test))
```

### Multiple Linear Regression using Gradient Descent from scratch (Stochastic Gradient Descent)

```python
class MLRSStochasticGradientDescent:
    def __init__(self,learning_rate,epoch):
        self._coef=None
        self._intercept=None
        self.learning_rate=learning_rate
        self.epoch=epoch
        
    def fit(self,X_train,y_train):
        self._intercept=0
        self._coef=np.ones(X_train.shape[1]) # we initilize an array of ones with shape of X_train as there will be that many betas as columns in X_train 
        for i in range(self.epoch):
            for j in range(X_train.shape[0]):
                random=np.random.randint(0,X_train.shape[0])
                y_pred=np.dot(X_train[random],self._coef)+self._intercept # X.Beta+intercept 
                slope_intercept=-2 * (y_train[random]-y_pred)
                self._intercept=self._intercept-(self.learning_rate*slope_intercept)

                slope_coef= -2*np.dot((y_train[random]-y_pred),X_train[random])
                self._coef=self._coef-(self.learning_rate*slope_coef)
        print(self._coef,self._intercept)
    def predict(self,X):
        return np.dot(X,self._coef)+self._intercept
```

```python
sgd=MLRSStochasticGradientDescent(learning_rate=0.01,epoch=100)
sgd.fit(X_train,y_train)
```

```python
y_pred=sgd.predict(X_test)
r2_score(y_test,y_pred)
```

### Multiple Linear Regression using Gradient Descent from scratch (Mini batch Gradient Descent)

```python
import random
class MLRMiniBatchGradientDescent:
    def __init__(self,learning_rate,epoch,batch_size):
        self._coef=None
        self._intercept=None
        self.learning_rate=learning_rate
        self.epoch=epoch
        self.batch_size=batch_size
        
    def fit(self,X_train,y_train):
        if self.batch_size<X_train.shape[0]:
            self._intercept=0
            self._coef=np.ones(X_train.shape[1]) # we initilize an array of ones with shape of X_train as there will be that many betas as columns in X_train 
            for i in range(self.epoch):
                for j in range(int(X_train.shape[0]/self.batch_size)):
                    random_indices=random.sample(range(X_train.shape[0]),self.batch_size)
                    y_pred=np.dot(X_train[random_indices],self._coef)+self._intercept # X.Beta+intercept 
                    slope_intercept=-2 * np.mean(y_train[random_indices]-y_pred)
                    self._intercept=self._intercept-(self.learning_rate*slope_intercept)

                    slope_coef= -2*np.dot((y_train[random_indices]-y_pred),X_train[random_indices])/self.batch_size
                    self._coef=self._coef-(self.learning_rate*slope_coef)
            print(self._coef,self._intercept)
    def predict(self,X):
        return np.dot(X,self._coef)+self._intercept
```

```python
mb=MLRMiniBatchGradientDescent(learning_rate=0.01,epoch=250,batch_size=int(X_train.shape[0]/70))
```

```python
mb.fit(X_train,y_train)
```

```python
r2_score(y_test,mb.predict(X_test))
```
